```python
from pyqpanda import *
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline


def draw(prog, name="f.png"):
    draw_qprog(prog, "pic", filename=name)
    plt.imshow(plt.imread(name))
    plt.axis("off")
    plt.show()


np.random.seed(0)
a = 1
b = 1
c = 100
M = 5
N = 5
m = 3
n = 2
A = 3
B = 3
alpha = abs(np.random.randn(M + 1, N + 1))
beta = abs(np.random.randn(M + 1, N + 1))
betai = 1
gammai = 1


def getH(*args):
    re = PauliOperator("", 1)
    for i in args:
        re *= PauliOperator({"": 0.5, "Z" + str(i): -0.5})
    return re


def getHp(a, b, c, M, N, m, n, A, B, alpha, beta):
    re = PauliOperator()
    for i in range(1, M + 1):
        for j in range(1, N + 1):
            re += (-2 * a * A * alpha[i, j] - 2 * b * B * beta[i, j]) * getH(
                i - 1, j - 1 + M
            )
            for k in range(1, M + 1):
                for l in range(1, N + 1):
                    re += (
                        a * alpha[i, j] * alpha[k, l] + b * beta[i, j] * beta[k, l]
                    ) * getH(i - 1, j - 1 + M, k - 1, l - 1 + M)
    for i in range(1, M + 1):
        re += (-2 * c * m) * getH(i - 1)
        for k in range(1, M + 1):
            re += c * getH(i - 1, k - 1)
    for j in range(N + 1):
        re += (-2 * c * n) * getH(j - 1 + M)
        for l in range(1, N + 1):
            re += c * getH(j - 1 + M, l - 1 + M)
    re.reduce_duplicates()
    return re
```

```python
def sumCNOT(a):
    re = 0
    for i in a:
        re += (len(i) - 1) * 2
    return re


def delCNOT(a):
    re = 0
    for i in range(len(a) - 1):
        x = a[i]
        y = a[i + 1]
        if x[-1] != y[-1]:
            continue
        for j in range(min(len(x), len(y)) - 1):
            if x[j] == y[j]:
                re += 1
            else:
                break
    return re * 2


def display(a):
    if isinstance(a, list):
        print(
            "总长度：",
            len(a),
            "\t原本的CNOT门数：",
            sumCNOT(a),
            "\t减少的CNOT门数：",
            delCNOT(a),
        )
        for i in a:
            display(i)
        return
    print("(", end="")
    for i in range(len(a) - 1):
        print(a[i], ",", end="", sep="")
    print(a[-1], ")", end="", sep="")


def order4():
    L = []
    # (xi,xj,yk,yl)
    for l in range(1, 6):
        for i in range(1, 6):
            for j in range(i + 1, 6):
                for k in range(l + 1, 6):
                    L.append(("x" + str(i), "x" + str(j), "y" + str(k), "y" + str(l)))
    return L


def order31():
    L0 = order4()
    L = [("x1", "x2", "y1")]
    for i in L0:
        if not (i[0] == L[-1][0] and i[1] == L[-1][1] and i[-1] == L[-1][-1]):
            L.append((i[0], i[1], i[-1]))
        L.append(i)
    return L


def order32():
    L = []
    # (yi,yj,xk):
    for k in range(1, 6):
        for i in range(1, 6):
            for j in range(i + 1, 6):
                L.append(("y" + str(i), "y" + str(j), "x" + str(k)))
    return L


def order21():
    L = []
    # (xi,xj) (xi,y5,xj)
    for j in range(1, 6):
        for i in range(j + 1, 6):
            L.append(("x" + str(i), "x" + str(j)))
            L.append(("x" + str(i), "y5", "x" + str(j)))
    return L


def order22():
    L0 = order32()
    L = [("y1", "x1")]
    for i in L0:
        if not (i[0] == L[-1][0] and i[-1] == L[-1][-1]):
            L.append((i[0], i[-1]))
        L.append(i)
    # (y5,xi)
    for i in range(1, 6):
        L.append(("y5", "x" + str(i)))
    return L


def order2():
    L = []
    # (yi,yj)
    for j in range(1, 6):
        for i in range(j + 1, 6):
            L.append(("y" + str(i), "y" + str(j)))
    L += order21()
    L += order22()
    return L


def order():
    return order2() + order31()


def finalOrder():
    L0 = order()
    L = []
    L.append((-1,))
    for i in L0:
        j = list(i)
        for idx, val in enumerate(j):
            if val[0] == "x":
                j[idx] = int(val[1]) - 1
            else:
                j[idx] = int(val[1]) + M - 1
        L.append(tuple(j))
    L.append((-1,))
    return L
```

```python
def getLambda(a, L):
    x = list(a)
    x.sort()
    # print(x)
    for i in L:
        if x == [y for idx, y in enumerate(i[0])]:
            return i[1]
    return 0
```

```python
machine = CPUQVM()
machine.init_qvm()
q = machine.qAlloc_many(10)
c = machine.cAlloc_many(10)
```

```python
def xianluyouhua():
    prog = QProg()

    for i in range(1, len(fo) - 1):
        pre = fo[i - 1]
        now = fo[i]
        nxt = fo[i + 1]

        start = 0
        if pre[-1] == now[-1]:
            while start < len(now) - 1 and pre[start] == now[start]:
                start += 1
        for j in range(start, len(now) - 1):
            prog << CNOT(q[now[j]], q[now[-1]])

        t = gammai * getLambda(now, L)
        prog << U3(q[now[-1]], 0, t, t)

        start = 0
        if now[-1] == nxt[-1]:
            while start < len(now) - 1 and now[start] == nxt[start]:
                start += 1
        for j in range(len(now) - 2, start - 1, -1):
            prog << CNOT(q[now[j]], q[now[-1]])

    for i in range(10):
        t = gammai * getLambda((i,), L)
        prog << U3(q[i], 0, t, t)
        prog << U3(q[i], 2 * betai, -np.pi / 2, np.pi / 2)

    return prog
```
