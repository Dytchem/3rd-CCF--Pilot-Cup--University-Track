```python
import matplotlib.pyplot as plt
import numpy as np
from pyqpanda import *
from scipy.stats import unitary_group

%matplotlib inline


def draw(prog):
    draw_qprog(prog, "pic", filename="a.png")
    plt.imshow(plt.imread("a.png"))
    plt.axis("off")
    plt.show()
```

```python
machine = CPUQVM()
machine.init_qvm()

q = machine.qAlloc_many(10)

prog = QProg()

(
    prog
    << CNOT(q[0], q[5])
    << CNOT(q[1], q[5])
    << CNOT(q[6], q[5])
    << U3(q[5], 1, 1, 1)
    << CNOT(q[6], q[5])
)

(
    prog
    << CNOT(q[7], q[5])
    << U3(q[5], 1, 1, 1)
    << CNOT(q[7], q[5])
    << CNOT(q[1], q[5])
    << CNOT(q[0], q[5])
)


draw(prog)
```
